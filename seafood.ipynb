```python
area = 'Moschato'
section = 'food'
category = '4bf58dd8d48988d1ce941735'    # seafood
limit = 100
radius = 1000

url3 = 'https://api.foursquare.com/v2/venues/explore?\
client_id={}&client_secret={}&near={}&v={}&\
section={}&categoryId={}&radius={}&limit={}'\
.format(CLIENT_ID, CLIENT_SECRET, area, VERSION, section, category, radius, limit)

url3 = url3.replace(" ", "%20")
url3 = url3.replace(",", "%2C")

url3
```

```python
resA = requests.get(url1).json()

venues1 = resA['response']['venues']

df1 = pd.json_normalize(venues1)
df1
```

```python
resB = requests.get(url2).json()

venues2 = resB['response']['venues']

df2 = pd.json_normalize(venues2)
df2
```

```python
resC = requests.get(url3).json()

venues3 = resC['response']['groups'][0]['items']

df3 = pd.json_normalize(venues3)
df3
```

```python
accepted_keywords = ['Mediterranean', 'Greek', 'Cretan', 'Meze', 'Tsipouro', 'Seafood', 'Fish', 'Magirio']

for i, row in df3.iterrows():
    
    cat = row['venue.categories'][0]['name']

    accepted = any(accepted_keyword in cat for accepted_keyword in accepted_keywords)

    if not accepted:
        df3.drop([i], inplace=True)


df = df3[['venue.id', 'venue.name', 'venue.location.formattedAddress']]        
df.reset_index(drop=True)
df
```

```python
# generate map centered at our location
venues_map = folium.Map(location=[latitude, longitude], zoom_start=13)

# add a red marker for Moschato
folium.CircleMarker(
    [latitude, longitude],
    radius = 10,
    color = 'red',
    fill = True,
    fill_color = 'red',
    fill_opacity = 0
).add_to(venues_map)

# add venues
for index, row in df3.iterrows():
    folium.CircleMarker(
        [row['venue.location.lat'], row['venue.location.lng']],
        radius = 3,
        color = 'blue',
        fill = True,
        fill_color = 'blue',
        fill_opacity = 1
    ).add_to(venues_map)
        
venues_map
```
