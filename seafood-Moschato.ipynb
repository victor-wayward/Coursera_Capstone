```python
# Get Moschato geolocation to be used for centering the map
address = 'Moschato'
geolocator = Nominatim(user_agent="foursquare_agent")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude

print(latitude, longitude)
```

We will create an 'explore' API request to Foursquare.com for restaurants under the section 'food' and category 'seafood'. We will request the hit limit to be 100 in order to override the default (30).

```python
area = 'Moschato'
section = 'food'
category = '4bf58dd8d48988d1ce941735'    # seafood category id
limit = 100

url = 'https://api.foursquare.com/v2/venues/explore?\
client_id={}&\
client_secret={}&\
near={}&v={}&\
section={}&\
categoryId={}&\
limit={}'\
.format(CLIENT_ID, CLIENT_SECRET, area, VERSION, section, category, limit)

url
```

```python
results = requests.get(url).json()
venues = results['response']['groups'][0]['items']
df = pd.json_normalize(venues)

df
```

After carefully inspecting the results, there are places that do not really fit in. These are coffee shops or restaurants that mainly serve meat plates. To clean up the resulting set we will create a set of accepted keywords that should appear under the 'venue.categories' column. 

```python
# Clean up results
accepted_keywords = ['Mediterranean', 'Greek', 'Cretan', 'Meze', 'Tsipouro', 'Seafood', 'Fish', 'Magirio']
for i, row in df.iterrows():
    categ = row['venue.categories'][0]['name']
    accepted = any(accepted_keyword in categ for accepted_keyword in accepted_keywords)
    if not accepted:
        df.drop([i], inplace=True)
        
# Reindex rows after clean up
df.reset_index(drop=True, inplace=True)

# Rename columns
df.rename(columns = {'venue.id':'id', 'venue.name':'name', 'venue.location.formattedAddress':'address', 'venue.location.lat':'lat', 'venue.location.lng':'lng'}, inplace = True) 

# Select only columns of interest
dff = df[['id', 'name', 'address', 'lat', 'lng']]

dff
```

Now, let's assume we have an index of how popular a place is. Certainly, our new restaurant does not have many chances if it is too close to a well established one. One way of adding scores would be by combining Google Maps score however for now, we will just generate a random number for each venue.

```python
df['score'] = np.random.randint(0, 500, dff.shape[0]) / 100

dff
```

For our map we will use red markers for restaurants that have a rating greater than 4/5. Also, we will use opacity to show areas that are covered by more than one restaurants.

```python
# generate map centered to our location
venues_map = folium.Map(location=[latitude, longitude], zoom_start=14)

# add venues
for i, row in dff.iterrows():

    strength = round(row['score'] / 5, 2)
    if strength > 0.8:
        mcolor = "Crimson"
    else:
        mcolor = "DarkSlateBlue"

    folium.CircleMarker(
        [row['lat'], row['lng']],
        radius = 50,
        color = "#CCC",
        popup = str(i) + ': ' + str(row['score']),
        fill = True,
        fill_color = mcolor,
        fill_opacity = strength,
        opacity = 1
    ).add_to(venues_map)

# add a red marker for Moschato
folium.CircleMarker(
    [latitude, longitude],
    radius = 10,
    color = 'white',
    fill = True,
    fill_color = 'red',
    fill_opacity = 1
).add_to(venues_map)
        
venues_map
```

By examining the map we can easily identify potential areas for our new seafood reastaurant in Moschato.
